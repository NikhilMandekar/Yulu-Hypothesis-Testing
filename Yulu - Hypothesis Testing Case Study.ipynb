### Hypothesis testing

#### Working Day has effect on number of electric cycles rented

- For the following question we will use a **Two sample t-test** since we are comparing the dependent variable based on two independent groups - Working days and Non Working days

- For the hypothesis testing we will be setting the **Confidence Interval** at **95%**.
- Significance level is **5%**.

Setting the null and alternate hypothesis for the problem statement.

- H0 = There is no difference on the number of cycles rented on working days and non working days.
- Ha = There is a difference in the number of cycles rented on working days and non working days



```python
from scipy import stats
alpha = 0.05 # for 95% confidence interval
```

```python
workday = df[df['workingday'] == 1]['count']
non_workday = df[df['workingday'] == 0]['count']
```

```python
# Since we are using 2 sample t test we assume that the two independent groups have equal variance.

# checking the same by using Levene's test 
levene_workday = stats.levene(workday, non_workday)
```

The null and alternate hypothesis for levene's statistic is as follows:
- H0 = The groups have equal variance
- Ha = The groups do not have equal variance

```python
# Checking the pvalue for levene's statistic

print(levene_workday.pvalue)
```

At significance level = 0.05 we can say that the **p-value is greater than alpha**, hence we ***fail to reject the null hypothesis***.

```python
# Two Sample t-test
t_stat, p_value = stats.ttest_ind(workday, non_workday, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)

if p_value < alpha:
    print("Decision: Reject H0")
else:
    print("Decision: Fail to Reject H0")
    
```

There is **no statistically significant difference** in the average number of electric cycles rented between working days and non-working days.

### 2 - No. of cycles rented similar or different in different seasons

```python
spring = df[df['season'] == 1]['count']
summer = df[df['season'] == 2]['count']
fall   = df[df['season'] == 3]['count']
winter = df[df['season'] == 4]['count']
```

```python
# checking the variance by using Levene's test 
levene_season = stats.levene(spring,summer,fall,winter )
```

```python
print(levene_season.pvalue)
```

```python
if (levene_season.pvalue < alpha):
    print ('Fail to reject null hypothesis')
else:
    print('Reject null hypothesis')
           
```

- All groups have **equal** variance.

- For this test we will be using **One way ANOVA** test since we are testing the effect of one independent variable (season) with 4 groups on one dependent variable

Setting the null and alternate hypothesis for the problem statement.

- H0 = The mean number of electric cycles rented is the same across all seasons.
- Ha = There will be atleast one season where the mean number of cycles rented is different.



```python
f_stat, p_value = stats.f_oneway(spring, summer, fall, winter)

print("test-statistic:", f_stat)
print("p-value:", p_value)
```

- As we have already set the significance value alpha as 0.05

```python
if p_value < 0.05:
    print("Reject H0")
    print("Atleast one group have different mean")
else:
    print("Fail to reject H0")
    print("All groups have same mean")
```

- The average number of electric cycles rented **differs significantly** across seasons.

### No. of cycles rented similar or different in different weather

- For this test we will be using **One way ANOVA** test since we are testing the effect of one independent variable (season) with 4 groups on one dependent variable

```python
clear = df[df['weather'] == 1]['count']
mild_extreme = df[df['weather'] == 2]['count']
mid_extreme = df[df['weather'] == 3]['count']
extreme = df[df['weather'] == 4]['count']
```

```python
# Levene's test for variance

levene_weather = stats.levene(clear,mild_extreme,mid_extreme,extreme)
print(levene_weather.pvalue)
```

```python
if (levene_weather.pvalue < alpha):
    print ('Fail to reject null hypothesis')
else:
    print('Reject null hypothesis')
           
```

- All groups have **equal** variance.

Setting the null and alternate hypothesis for the problem statement.

- H0 = The mean number of electric cycles rented is the same across all weather conditions.
- Ha = There will be atleast one weather where the mean number of cycles rented is different.



```python
f_stat, p_value = stats.f_oneway(clear,mild_extreme,mid_extreme,extreme)

print("test-statistic:", f_stat)
print("p-value:", p_value)
```

- As we have already set the significance value alpha as 0.05

```python
if p_value < 0.05:
    print("Reject H0")
    print("Atleast one group have different mean")
else:
    print("Fail to reject H0")
    print("All groups have same mean")
```

- The **weather condition** clearly has a direct effect on the number of cycles rented.

### Weather is dependent on season

- For checking the dependency of the two categorical variables we will use **chi-square test of independence**.

Setting the null and alternate hypothesis for the problem statement.

- H0 = Weather and Season variables are independent.
- Ha = Weather and Season variables are dependent.



```python
contingency_table = pd.crosstab(df['season'], df['weather']) 
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table) 
print("Chi-square statistic:", chi2)
print("Degrees of freedom:", dof) 
print("p-value:", p_value) 
```

```python
if p_value < alpha: 
    print("Reject H0") 
else: 
    print("Fail to Reject H0")
```

- As we can see here since the p-value for the chi square test is **less** than alpha we reject the null hypothesis meaning that the variables **Weather** and **Season** are dependent.

The results show that weather and season are closely related, which means bike rental demand is influenced by seasonal weather changes. This insight can help the business prepare in advance for demand variations throughout the year.
